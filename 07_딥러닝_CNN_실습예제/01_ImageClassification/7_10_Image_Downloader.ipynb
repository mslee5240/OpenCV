    - 입력 : 몇세트를 다운로드할지, 검색이미지 키워드, 다운로드할 이미지 개수
    - 다운로드 : 현재 디렉토리 > data 폴더 > 검색키워드 폴더 하위에 다운로드

```python
# code here
# 다운로드할 이미지 키워드와 개수 지정하여 반복 다운로드하기
from bing_image_downloader import downloader
import os

# 몇 세트를 받을지 먼저 입력
num_sets = int(input("몇개의 이미지 세트를 다운로드 할까요? => "))

# 현재 디렉토리의 절대 경로 가져오기
current_directory = os.path.abspath('.')

for _ in range(num_sets):
    # 키워드와 개수 입력
    keywords = input("검색할 이미지 키워드를 입력하세요 (예: 노란 바나나) => ")
    limit = int(input("다운로드할 이미지의 개수를 입력하세요 =>  "))

    # 다운로드 실행
    downloader.download(
        keywords,
        limit=limit,
        output_dir=current_directory,
        adult_filter_off=True,
        force_replace=False,
        timeout=60
    )
```

```python
# high_res_download_fixed.py
from bing_image_downloader import downloader
import os

# 1) 라벨 목록 정의
labels = ['unripe', 'ripe', 'overripe']
limit = 50  # 각 클래스당 다운로드할 이미지 수

# 2) 상위 data/ 폴더 생성
base_dir = os.path.abspath('.')
data_dir = os.path.join(base_dir, 'data')
os.makedirs(data_dir, exist_ok=True)

# 3) 라벨별 하위 폴더 생성 & 다운로드
for label in labels:
    out_dir = os.path.join(data_dir, label)
    os.makedirs(out_dir, exist_ok=True)

    # "unripe banana", "ripe banana", "overripe banana" 식으로 검색
    query = f"{label} banana"

    downloader.download(
        query=query,             # 검색어
        limit=limit,             # 이미지 개수
        output_dir=out_dir,      # data/<label>/ 폴더에 저장
        adult_filter_off=True,   # 성인 필터 해제
        force_replace=False,     # 이미 있으면 덮어쓰기 안 함
        timeout=60,              # 연결 타임아웃(초)
    )

    print(f"[✓] '{query}' 에 대해 최대 {limit}장의 고해상도 사진을 '{out_dir}'에 다운로드했습니다.")
```
